Задание 1 Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings.head(1)
```

```python
movies.head(1)
```

```python
"склеиваем 2 датафрейма"
joined = ratings.merge(movies, on='movieId', how='left')
"подсчитываем в неи по рейтингу 5.0"
top_movie_id = joined[joined['rating'] == 5.0]['movieId'].value_counts().head(1).index[0]
"выведем полную ин-фу по этому фильму"
print(top_movie_info := movies[movies['movieId'] == top_movie_id])
```

Задание 2 По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. 
Не учитывайте в расчётах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head(1)
```

```python
"поисковик по странам"
power[ power['country'].str.contains('est', case=False) ]['country'].unique()
```

```python
"по странам"
power = power[ (power['country']=='Latvia') | (power['country']=='Lithuania')| (power['country']=='Estonia')]
"по годам"
years = power[(power['year']>=2005)&(power['year']<=2010)&(power['quantity']>0)]
"по категориям"
categories = years[years.category.isin([4,12,21])]
"сумма"
total_quantity = categories['quantity'].sum()

print(F'Суммарное потребление: {total_quantity} Гвт')
```
