```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head(1)
```

```python
def determine_source_type(row):
    if row['traffic_source'].lower() in ['yandex', 'google']:
        return 'organic'#если источник traffic_source равен Yandex или Google, то в source_type ставится organic
    elif row['traffic_source'] in ['paid', 'email']:
        if row['region'] == 'Russia':  
            return 'ad'#для источников paid и email из России ставим ad
        else:
            return 'other'#для источников paid и email не из России ставим other
    else:
        return row['traffic_source']#все остальные варианты берём из traffic_source без изменений


log['source_type'] = log.apply(determine_source_type, axis=1)
#log.loc[log.source_type == 'organic'].head()
log.head(10)
```

```python
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
user_counts = ratings.groupby('userId').size()
```

```python
# Фильтрация пользователей с более чем 100 оценками
active_users = user_counts[user_counts > 100].index
```

```python
active_users_timestamps = ratings[ratings['userId'].isin(active_users)]
```

```python
lifetime = active_users_timestamps.groupby('userId')['timestamp'].agg(['max', 'min'])
lifetime['duration'] = lifetime['max'] - lifetime['min']
```

```python
average_lifetime = lifetime['duration'].mean()
print(f'Cреднее время жизни пользователей, которые выставили более 100 оценок {average_lifetime}')
```
