Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
rating = pd.read_csv('ratings.csv')
```

```python
def watcher(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'


rating['class'] = rating['rating'].apply(watcher)
full_rates = rating.merge(movies, on='movieId', how='left')
full_rates.head(10)
```

Задание 2 Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
keywords_geo = pd.read_csv('keywords.csv', encoding='UTF-8')
keywords_geo.head(3)
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}


def classify_region(row):
    keyword_set = row['keyword'].lower()  
    for region, city_list in geo_data.items():
        for city in city_list:
            if city in keyword_set:  
                return region
    return 'undefined'  

"применяем функцию"
keywords_geo['region'] = keywords_geo.apply(classify_region, axis=1)

"сортируем по региону центр,дальний восток для проверки"
#center_keywords = keywords_geo[keywords_geo['region'] == 'Центр']
#center_keywords.head(3)
vostok_keywords = keywords_geo[keywords_geo['region'] == 'Дальний Восток']
vostok_keywords.head(3)
```

В переменную years запишите список из всех годов с 1950 по 2010 года.

```python
years = pd.date_range(start='1950-01-01', end='2010-12-31', freq='YE').year.tolist()
```

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте алгоритм:

для каждой строки пройдите по всем годам списка years;
если номер года присутствует в названии фильма, то функция возвращает этот год, как год выпуска;
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.

```python
def production_year(title):
    if isinstance(title, str):
        for year in years:
            if str(year) in title:
                return year
    return 1900 

full_rates['year'] = full_rates['title'].apply(production_year)

full_rates.head(3)
    
```

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
average_ratings = full_rates.groupby('year')['rating'].mean().reset_index()

average_ratings.columns = ['год', 'средняя оценка']


sorted_average_ratings = average_ratings.sort_values(by='средняя оценка', ascending=False)


sorted_average_ratings.head(5)
```
